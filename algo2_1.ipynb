```python
import networkx as nx
import random 
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

### Numero di vertici

```python
n=20
```

### Probabilit√†

```python
p=10
```

### Creo i vertici

```python
for i in range(n):
    G.add_node(i)
```

### Creo archi

```python
for i in range(n):
    for j in range(n):
        a=random.randint(0,100)
        if i!=j and a<p:
            G.add_edge(i,j)
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```
